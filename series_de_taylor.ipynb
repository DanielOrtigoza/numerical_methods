# Ejemplo 1

La serie de taylor de la funciÃ³n $\dfrac{1}{1 - x}$ se puede expresar como:

$$ \dfrac{1}{1 - x} = \sum_{n = 0}^{\infty} x ^n $$

```python
def ejex_fun1(x: int, n: int) -> list:

    npuntos = n
    xmin = -x
    xmax = x
    deltax = (xmax - xmin) / npuntos
    
    vector = []

    for i in range(npuntos):
        vector.append(xmin + (deltax * i))

    return vector
```

```python
def fun1(a: int, x: list) -> list:

    fun = []

    for i in range(len(x)):
        fun.append(a / (1 - x[i]))
    return fun
```

```python
def serie_de_taylor_fun1(a: int, x: int, n: int, k: int) -> list:

    serie_taylor = []
    x_axis = ejex_fun1(x, n)

    for i in range(n):
        y = 0
        for j in range(k + 1):
            y = y + (a * (x_axis[i] ** j))
        serie_taylor.append(y)

    return serie_taylor
```

```python
a = 1
x = 0.9
n = 100
k = 5

x_plot_ex1 = ejex_fun1(x, n)
y_plot_ex1 = fun1(a, x_plot_ex1)

taylor_ex1 = serie_de_taylor_fun1(a, x, n, k)
```

```python
%matplotlib widget

plt.style.use('fivethirtyeight')

plt.figure(1, figsize = (cm2inch(20), cm2inch(13)), constrained_layout = True) # [10,5] significa el tamaÃ±o de la figura [x, y]

plt.plot(x_plot_ex1, y_plot_ex1, '-', color = colors['green'], linewidth = 2.5, label = '$f(x) = \dfrac{a}{1-x}$')
plt.plot(x_plot_ex1, taylor_ex1, '--', color = colors['blue'], linewidth = 2.5, label = 'AproximaciÃ³n serie de Taylor de grado {}'.format(k))

plt.ylabel('$y$', fontsize = 'large')
plt.xlabel('$x$', fontsize = 'large')
plt.title('Serie de Taylor de $f(x) = \dfrac{a}{1-x}$', fontsize = 'large')
plt.legend(framealpha = 0.8, fontsize = 'large')

plt.grid(visible = True, which = 'major', color = '#666666', linestyle = '-', alpha = 0.6)
plt.minorticks_on()
plt.grid(visible = True, which = 'minor', color = '#999999', linestyle = '-', alpha = 0.2)
plt.tight_layout()

# plt.savefig('ajuste_lineal_punto_1_taller_2.jpg', dpi = 1200, quality = 95)

plt.show()
```

# Ejemplo 2

La serie de taylor de la funciÃ³n $e^x$ se puede expresar como:

$$ e^x = \sum_{n = 0}^{\infty} \dfrac{x ^n}{n!} $$

```python
def factorial_fun2(x: int) -> int:

    fac = 1

    for i in range(1, x + 1):
        fac = fac * i
    return fac
```

```python
def ejex_fun2(xmin: int, xmax: int, npuntos: int) -> list:

    n = npuntos
    x1 = xmin
    x2 = xmax
    deltax = (x2 - x1) / n
    
    vector = []

    for i in range(n):
        vector.append(xmin + (deltax * i))

    return vector
```

```python
def fun2(x: list) -> list:

    fun = []

    for i in range(len(x)):
        fun.append(math.exp(x[i]))

    return fun
```

```python
def serie_de_taylor_fun2(xmin: int, xmax: int, npuntos: int, k: int) -> list:

    serie_taylor = []
    x_axis = ejex_fun2(xmin, xmax, npuntos)

    for i in range(npuntos):
        y = 0
        for j in range(k + 1):
            y = y + ((x_axis[i] ** j) / factorial_fun2(j))
        serie_taylor.append(y)

    return serie_taylor
    
```

```python
xmin = 0
xmax = 20
n = 1000
k = 20

x_plot_ex2 = ejex_fun2(xmin, xmax, n)
y_plot_ex2 = fun2(x_plot_ex2)

taylor_ex2 = serie_de_taylor_fun2(xmin, xmax, n, k)
```

```python
%matplotlib widget

plt.style.use('fivethirtyeight')

plt.figure(2, figsize = (cm2inch(20), cm2inch(13)), constrained_layout = True) # [10,5] significa el tamaÃ±o de la figura [x, y]

plt.plot(x_plot_ex2, y_plot_ex2, '-', color = colors['green'], linewidth = 2.5, label = '$f(x) = e^x$')
plt.plot(x_plot_ex2, taylor_ex2, '--', color = colors['blue'], linewidth = 2.5, label = 'AproximaciÃ³n serie de Taylor de grado {}'.format(k))

plt.ylabel('$y$', fontsize = 'large')
plt.xlabel('$x$', fontsize = 'large')
plt.title('Serie de Taylor de $e^x$', fontsize = 'large')
plt.legend(framealpha = 0.8, fontsize = 'large')

plt.grid(visible = True, which = 'major', color = '#666666', linestyle = '-', alpha = 0.6)
plt.minorticks_on()
plt.grid(visible = True, which = 'minor', color = '#999999', linestyle = '-', alpha = 0.2)
plt.tight_layout()

# plt.savefig('ajuste_lineal_punto_1_taller_2.jpg', dpi = 1200, quality = 95)

plt.show()
```

# Ejemplo 3

La serie de taylor de la funciÃ³n $sin(x)$ se puede expresar como:

$$ \sin(x) = \sum_{n = 0}^{\infty} \dfrac{(-1)^n}{(2n+1)!} x^{2n+1} $$

```python
def factorial_fun3(x: int) -> int:

    fac = 1

    for i in range(1, x + 1):
        fac = fac * i
    return fac
```

```python
def ejex_fun3(xmin: int, xmax: int, npuntos: int) -> list:

    n = npuntos
    x1 = xmin
    x2 = xmax
    deltax = (x2 - x1) / n
    
    vector = []

    for i in range(n):
        vector.append(xmin + (deltax * i))

    return vector
```

```python
def fun3(x: list) -> list:

    fun = []

    for i in range(len(x)):
        fun.append(math.sin(x[i]))

    return fun
```

```python
def serie_de_taylor_fun3(xmin: int, xmax: int, npuntos: int, k: int) -> list:

    serie_taylor = []
    x_axis = ejex_fun3(xmin, xmax, npuntos)

    for i in range(npuntos):
        y = 0
        for j in range(k + 1):
            y = y + ((((-1) ** j) / (factorial_fun3((2*j) + 1))) * (x_axis[i] ** ((2*j) + 1)))
        serie_taylor.append(y)

    return serie_taylor
```

```python
xmin = - 10
xmax = 10
npuntos = 100
k = 10

x_plot_ex3 = ejex_fun3(xmin, xmax, npuntos)
y_plot_ex3 = fun3(x_plot_ex3)

taylor_ex3 = serie_de_taylor_fun3(xmin, xmax, npuntos, k)
```

```python
%matplotlib widget

plt.style.use('fivethirtyeight')

plt.figure(3, figsize = (cm2inch(20), cm2inch(13)), constrained_layout = True) # [10,5] significa el tamaÃ±o de la figura [x, y]

plt.plot(x_plot_ex3, y_plot_ex3, '-', color = colors['aqua'], linewidth = 2.5, label = '$f(x) = \sin(x)$')
plt.plot(x_plot_ex3, taylor_ex3, '--', color = colors['purple'], linewidth = 2.5, label = 'AproximaciÃ³n serie de Taylor de grado {}'.format(k))

plt.ylabel('$y$', fontsize = 'large')
plt.xlabel('$x$', fontsize = 'large')
plt.title('Serie de Taylor de $\sin(x)$', fontsize = 'large')
plt.legend(framealpha = 0.8, fontsize = 'large')

plt.grid(visible = True, which = 'major', color = '#666666', linestyle = '-', alpha = 0.6)
plt.minorticks_on()
plt.grid(visible = True, which = 'minor', color = '#999999', linestyle = '-', alpha = 0.2)
plt.tight_layout()

# plt.savefig('ajuste_lineal_punto_1_taller_2.jpg', dpi = 1200, quality = 95)

plt.show()
```
